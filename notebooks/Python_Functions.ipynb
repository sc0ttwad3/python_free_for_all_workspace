```python
# function without a return statement in body, returns None
# "The nonlocal statement causes the listed identifiers to refer 
# to previously bound variables in the nearest
# enclosing scope excluding globals."
#
def outer():
    test = 1
    def inner():
        global test
        test = 2
        print('inner', test)
    
def outer():
    test = 1
    def inner():
        global test
        test = 2
        print('inner', test)
    inner()
    print('outer', test)
       
    
test = 0
outer()
print('global', test)
```

```python
# argument passing is just assigning an object to a local variable name
# inside a function this does not affect the caller
# changing a MUTABLE object argument in a function DOES affect the caller
# Within func, name y is created and points to same object as xArgs is
xArg = 3
def func(y):
    print(y)
    
func(xArg)

# When we call the function, we are telling runtime which objects the arguments names must point to
# Variable positional args
# Pass a variable number of positional arguments
def minimum(*n):
    # print(n) # n is a tuple
    if n: # explained after the code
        mn = n[0]
        for value in n[1:]:
            if value < mn:
                mn = value
        print(mn)

minimum(1,-3,-7,9)
minimum()
```
